```python
import torch
import torch.nn as nn
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
data, _ =datasets.make_blobs(centers=[[-0.5,-0.5],[0.5,0.05]],n_samples=1000,cluster_std=0.4)
```

```python
X=data[:,0]
Y=data[:,1]
```

```python
X=torch.Tensor(X).unsqueeze(1)
Y=torch.Tensor(Y).unsqueeze(1)
```

```python
model=nn.Linear(1,1)
```

```python
def plotModel(model):
    weights, bias=list(model.parameters())
    
    plt.scatter(X,Y)
    plt.plot([-2,2],[-2*weights+bias,2*weights+bias])
```

```python
criterium=nn.MSELoss()
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
for index in range(1000):
    y_pred=model(X)
    loss=criterium(y_pred, Y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    plotModel(model)
```
